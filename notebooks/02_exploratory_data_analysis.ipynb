## Load Data

This notebook explores cases, deaths and vaccination trends.

```python
import pandas as pd
import numpy as np
from pathlib import Path

DATA_PATH = Path("../data/cleaned/covid_intelligence_tableau_ready.csv")
df = pd.read_csv(DATA_PATH)
df.head()
```

## Top Countries by Infection Rate

Population-adjusted metrics are more useful than raw totals.

```python
latest = df.sort_values("date").groupby("location").tail(1)
latest[["location", "infection_rate_pct"]].sort_values("infection_rate_pct", ascending=False).head(10)
```

## Death Rate by Continent

Compare case fatality rates across continents.

```python
latest.groupby("continent")["case_fatality_rate_pct"].mean().sort_values(ascending=False)
```
