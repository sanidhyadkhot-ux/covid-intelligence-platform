## Load Dataset

Load the cleaned COVID intelligence dataset and inspect the first few rows.

```python
import pandas as pd
import numpy as np
from pathlib import Path

DATA_PATH = Path("../data/cleaned/covid_intelligence_tableau_ready.csv")
df = pd.read_csv(DATA_PATH)
df.head()
```

## Data Shape and Missing Values

Check row counts, column counts and missing values.

```python
print("Rows:", len(df))
print("Columns:", len(df.columns))
df.isna().sum().sort_values(ascending=False).head(20)
```

## Duplicate Check

Validate whether country-date level rows are unique.

```python
duplicate_rows = df.duplicated(subset=["location", "date"]).sum()
print("Duplicate country-date rows:", duplicate_rows)
```
