# 03 Feature Engineering

Master's-level portfolio notebook for the COVID Intelligence Platform 2026.

## Load Data

Create features used in Tableau and Streamlit dashboards.

```python
import pandas as pd
import numpy as np
from pathlib import Path

DATA_PATH = Path("../data/cleaned/covid_intelligence_tableau_ready.csv")
df = pd.read_csv(DATA_PATH)
df.head()
```

## KPI Features

Calculate infection rate, case fatality rate and vaccination coverage.

```python
df["infection_rate_pct_calc"] = (df["total_cases"] / df["population"] * 100).round(3)
df["case_fatality_rate_pct_calc"] = (df["total_deaths"] / df["total_cases"].replace(0, np.nan) * 100).fillna(0).round(3)
df["vaccination_coverage_pct_calc"] = (df["total_vaccinations"] / df["population"] * 100).clip(0, 220).round(2)
df[["location", "date", "infection_rate_pct_calc", "case_fatality_rate_pct_calc", "vaccination_coverage_pct_calc"]].head()
```

## Risk Score

Create a simple risk score for demonstration.

```python
df["risk_score_demo"] = (
    df["case_fatality_rate_pct_calc"] * 25
    + df["infection_rate_pct_calc"] * 0.8
    - df["vaccination_coverage_pct_calc"] * 0.05
).clip(0, 100).round(1)

df[["location", "date", "risk_score_demo"]].head()
```
