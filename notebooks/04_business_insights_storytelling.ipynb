## Load Data

This notebook converts analytics into manager-ready insights.

```python
import pandas as pd
import numpy as np
from pathlib import Path

DATA_PATH = Path("../data/cleaned/covid_intelligence_tableau_ready.csv")
df = pd.read_csv(DATA_PATH)
df.head()
```

## Insight 1

Raw case totals should not be used alone.

```python
latest = df.sort_values("date").groupby("location").tail(1)
top_infection = latest.sort_values("infection_rate_pct", ascending=False).iloc[0]
top_death = latest.sort_values("case_fatality_rate_pct", ascending=False).iloc[0]
top_vax = latest.sort_values("vaccination_coverage_pct", ascending=False).iloc[0]

print("Top infection rate:", top_infection["location"], top_infection["infection_rate_pct"])
print("Top death rate:", top_death["location"], top_death["case_fatality_rate_pct"])
print("Top vaccination coverage:", top_vax["location"], top_vax["vaccination_coverage_pct"])
```

## Executive Summary Output

Create a short summary suitable for README, LinkedIn or a dashboard.

```python
summary = f'''
Executive Summary:
- {top_infection["location"]} recorded the highest population-adjusted infection rate.
- {top_death["location"]} recorded the highest case fatality rate.
- {top_vax["location"]} recorded the highest vaccination coverage.
- The dashboard should focus on population-adjusted metrics, not raw case totals.
'''
print(summary)
```
